## Converting ProcessTime

```python
lx.pdf['ProcessTime'] = pd.to_datetime(lx.pdf['ProcessTime'],errors='coerce') # NaT
lx.pdf.dtypes
```

## Converting Value

```python
lx.pdf.Value = lx.pdf.Value.str.replace(',', '.')
```

```python
lx.pdf.Value=pd.to_numeric(lx.pdf.Value,errors='coerce') # NaN 
```

```python
df=lx.pdf[lx.pdf['ID'].str.contains('^TABL')]
```

# SIR 3S Processdata Import File

```python
dfs=df.sort_values(by=['ID','ProcessTime','#LogTime'])
```

```python
dfs['ID'].replace('(TABL~)(\S+)~~~SWVT', r'\2',regex=True,inplace=True)
```

```python
dfs['LogDay'], dfs['LogTime'] = dfs['#LogTime'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S.%f')[:-3]).str.split(' ',1).str
```

```python
dfs['ProcessDay'], dfs['ProcessTime'] = dfs['ProcessTime'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S.%f')[:-3]).str.split(' ',1).str
```

```python
dfs.to_csv('20171103__000001_TABL.csv',sep=';',index=False,columns=['ID','ProcessDay','ProcessTime','Value'])
```
